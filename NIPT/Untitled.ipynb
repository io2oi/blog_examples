```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

mu = 0.90
sigma = 0.001
# csigma = sigma * np.sqrt((100)
```

```python
fig = plt.figure()
pp = fig.add_subplot(1,1,1)

###  normal distribution 표시
tmpx = np.linspace(mu - 3*sigma, mu+3*sigma,100)
tmpy = stats.norm.pdf(tmpx, mu, sigma)
plt.plot(tmpx, tmpy)

### 정상일 경우의 가상 분포
pp.axvline(x=mu, color='r')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
pp.text(0.5,0.95,r'$\mu=%.2f, \sigma=%.3f$' %(mu, sigma), transform=pp.transAxes, fontsize=10,verticalalignment='top', bbox=props)

### T21_0
pp.axvline(x=0.899, color='b')
```
